```python
#3

import numpy as np
import matplotlib.pyplot as plt

def f(k, x, m):
    y = k*x + m
    return y

x = np.linspace(0, 10)

plt.plot(x, f(3, x, 1))
plt.title("En sådär kännande graf")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#4

import numpy as np
import matplotlib.pyplot as plt

def f(k, x, m):
    y = x**k + m
    return y

x = np.linspace(-2, 2)

plt.plot(x, f(2, x, -1))
plt.title("En lycklig och aldrig negativ graf")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
