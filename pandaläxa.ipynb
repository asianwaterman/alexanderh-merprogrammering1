```python
#1a

import pandas as pd
import numpy as np


df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

förvärvsarbetande = df[df["sysselsättning"] == "förvärvsarbetande"]

antal = förvärvsarbetande["2019"].sum(axis=0)

print(f"2019 var det {antal} förvärvsarbetande")








```

```python
#1b

import pandas as pd
import numpy as np

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

ej_förvärvsarbetande = df[df["sysselsättning"] == "ej förvärvsarbetande"]
antal1 = ej_förvärvsarbetande["2019"].sum(axis=0)

print(f"2019 var det {antal1} som ej förvärvsarbetade")

```

```python
#1c

import matplotlib.pyplot as plt

labels = "ej förvärvsarbetande", "förvärvsarbetande"
size = [antal, antal1]

fig1, ax1 = plt.subplots()
ax1.pie(size, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal') 
```

```python
#2a

import pandas as pd
import numpy as np

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

förvärvsarbetande = df[df["sysselsättning"] == "förvärvsarbetande"]

män_förvärvsarbetande = förvärvsarbetande[förvärvsarbetande["kön"] == "män"]

antalmän = män_förvärvsarbetande["2019"].sum(axis=0)
print(f"Antal arbetande män: {antalmän}")
```

```python
#2b

import pandas as pd
import numpy as np

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

förvärvsarbetande = df[df["sysselsättning"] == "förvärvsarbetande"]

kvinnor_förvärvsarbetande = förvärvsarbetande[förvärvsarbetande["kön"] == "kvinnor"]

antalkvinnor = kvinnor_förvärvsarbetande["2019"].sum(axis=0)
print(f"Antal arbetande kvinnor: {antalkvinnor}")
```
